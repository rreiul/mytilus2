```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Leer el archivo y pasarlo a DF
df = pd.read_csv("../data/raw/cv_pontevedra.csv")

# Mostrar Df
df
```

```python
df = df.drop(columns=['ria', 'zona', 'cuadricula', 'especie', 'psp', 'asp', 'bioenasio', 'aza', 'ytx', 'stx'])
df
```

```python
df_cangasb = df[df['punto'].str.contains('CANGAS-B')]
df_cangasb.loc[:, 'ao'] = df_cangasb['ao'].str.split(' ', n=1).str[0]
df_cangasb
```

```python
df_cangasb['ao'] = df_cangasb['ao'].replace('---', np.nan)
df_cangasb
```

```python
import pandas as pd
import numpy as np
import re

df2 = df_cangasb.dropna() 
```

```python
df2['ao'] = df2['ao'].replace('<', '0')
df2
```

```python
df2 = df2.rename(columns={'dia': 'fecha'})
df2['fecha'] = pd.to_datetime(df2['fecha'], format='%d/%m/%Y')
df2['ao'] = df2['ao'].str.replace(',', '.')
df2['ao'] = pd.to_numeric(df2['ao'])
df2 = df2.groupby('fecha')['ao'].mean()
df2.to_csv('../data/processed/medias2.csv')
```
