```python
import pandas as pd
from io import StringIO

# Cargar el dataset desde un archivo (puede ser .csv, .xlsx, etc.)
# Aquí asumimos que es un archivo .csv llamado 'dataset.csv'
df = pd.read_csv('../data/raw/variables.csv')

# Identificar las columnas que terminan en '_val'
val_columns = [col for col in df.columns if col.endswith('_val')]

# Filtrar las filas que no tienen 3, 4 o 9 en las columnas que terminan en '_val'
filtered_df = df[~df[val_columns].isin([3, 4, 9]).any(axis=1)]

# Mostrar el DataFrame resultante
filtered_df
```

```python
filtered_df = filtered_df.drop(columns=val_columns)
filtered_df
```

```python
for col in filtered_df.columns:
    if col != 'fecha':
        # Convertir la columna a string, luego reemplazar comas por puntos y manejar valores inválidos
        filtered_df[col] = filtered_df[col].astype(str).str.replace(',', '.').replace('-', 'NaN').astype(float)

filtered_df['fecha'] = pd.to_datetime(filtered_df['fecha'], format='%d/%m/%Y %H:%M:%S').dt.date

# Agrupar por la columna 'fecha' y calcular la media del resto de columnas
grouped_df = filtered_df.groupby('fecha').mean().reset_index()

# Mostrar el DataFrame resultante
print(grouped_df)
```
