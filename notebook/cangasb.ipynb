```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/raw/cangasb.csv")
df['fecha'] = pd.to_datetime(df['dia'], format="%d/%m/%Y")
```

```python
df = df.sort_values(by='fecha')

# Configura la figura y los ejes
fig, ax = plt.subplots(figsize=(18,6))

# Itera sobre cada fila del DataFrame
for index, row in df.iterrows():
    # Define los colores según el valor booleano
    color = 'red' if row['abierto/cerrado'] else 'white'
    
    # Dibuja un rectángulo coloreado para cada fila
    ax.axvspan(row['fecha'], df['fecha'].iloc[index+1] if index < len(df)-1 else df['fecha'].iloc[-1], color=color, alpha=0.5)

# Configura los ejes
ax.set_xlabel('Fecha')
ax.set_ylabel('Zona')
ax.set_title('Representación de zonas con booleanos')
# Muestra el gráfico
plt.show()
```
